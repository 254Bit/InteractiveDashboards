```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Population Distribution in 2024
plt.figure(figsize=(10, 6))
sns.histplot(df['Population 2024'], bins=30, kde=True)
plt.title('Population Distribution in 2024')
plt.xlabel('Population')
plt.ylabel('Frequency')
plt.show()

# Annual Population Change Distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['Annual Change'], bins=30, kde=True)
plt.title('Annual Population Change Distribution')
plt.xlabel('Annual Change (%)')
plt.ylabel('Frequency')
plt.show()
```
